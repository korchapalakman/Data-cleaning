# Dealing with Outliers

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Choose a mean,standard deviation, and number of samples

def create_ages(mu=50,sigma=13,num_samples=100,seed=42):

    # Set a random seed in the same cell as the random call to get the same values as us
    # We set seed to 42 (42 is an arbitrary choice from Hitchhiker's Guide to the Galaxy)
    np.random.seed(seed)

    sample_ages = np.random.normal(loc=mu,scale=sigma,size=num_samples)
    sample_ages = np.round(sample_ages,decimals=0)
    
    return sample_ages
```

```python
# create a normal distribution of ages

sample = create_ages()
sample
```

```python
# check the distribution 

sns.displot(sample, bins=20);
```

```python
# use another visualization 
# boxplot will show outliers


sns.boxplot(data=sample)
```

```python
ser = pd.Series(sample)
```

```python
ser.describe()
```

```python
# define IQR
# and the refion for outliers min -+1.5iqr
# everything below or above that is an outlier

# manual way

iqr = 55.25000 - 42.00000
lower_limit =42.0000- 1.5*iqr
lower_limit
```

```python
# we may later filter the data to 
# be free from outliers

ser[ser>lower_limit]

# notice that it has 1 value less
```

```python
# may do the above calculation automatically

q90, q95 = np.percentile(sample, [90,95])

# used tuple unpaking since it is an array 
```

```python
ser[ser<q95]
```

```python
q25, q75 = np.percentile(sample, [25,75])
```

```python
iqr = q75 - q25
```

```python
ser[ser<q75+1.5*iqr]
```
